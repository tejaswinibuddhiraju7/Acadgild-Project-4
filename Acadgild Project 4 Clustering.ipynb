### Problem 1 
### There are various stocks for which we have collected a data set, which all stocks are apparently similar in performance

```python
df['NEW_DATE'] = pd.to_datetime(df['DATE'],unit='s')
```

```python
df.drop(columns=['DATE','SP500'],axis=1,inplace=True)
```

```python
df.head(3) #Returns the first 3 rows of stocks dataframe
```

```python
df.set_index('NEW_DATE',inplace=True)
```

```python
df.head()
```

```python
df_transpose = df.transpose() #Creating transpose of the df dataframe
```

```python
df_transpose.head(3) #Returns the first 3 rows of df_transpose dataframe
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

pca_clf = PCA(n_components=3)
```

```python
pca_reduced = pca_clf.fit_transform(df_transpose)
```

```python
plt.scatter(pca_reduced[:,0],pca_reduced[:,1], alpha=0.5)
```

```python
clusters = []
inertia = []
for i in range(2,8):
    km=KMeans(n_clusters=i)
    km.fit(pca_reduced)
    clusters.append(i)
    inertia.append(km.inertia_)
```

```python
#Plot to find number of clusters (Elbow Method)
plt.figure(figsize=(10,8))
plt.plot(clusters,inertia)
plt.show()
```

```python
# We can see that eithe 5 or 6 is a good estimate of no_of_clusters, we have choosen 5 clusters
km = KMeans(n_clusters = 5)
km.fit(pca_reduced)
```

```python
#Predicted values using KMeans
predicted = km.predict(pca_reduced)
```

```python
#Scatter Plot
plt.figure(figsize=(10,8))
plt.scatter(pca_reduced[:,0],pca_reduced[:,1],c=predicted)
plt.show()
```

```python
df_transpose['Predicted'] = predicted
```

```python
df_transpose['Predicted'].value_counts()
```

```python
# Stocks which belongs to Cluster 4
df_transpose[df_transpose['Predicted']==4]
```
